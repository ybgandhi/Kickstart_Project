```python
# Dependencies
import pandas as pd
import json
from geojson import Feature, FeatureCollection, Point
```

```python
# ser url variable from where the library will scrape data from
url = 'https://sites.google.com/site/32nflteamsabsoluteandrelative/'
```

```python
scrape = pd.read_html(url)
scrape
```

```python
type(scrape)
```

```python
df = scrape[1]
df.head()
```

```python
# remove first row
df_drop_first_row = df.iloc[1:]
df_drop_first_row.head()
```

```python
# remove last row
df_drop_last_row = df_drop_first_row[:-1]
df_drop_last_row.head()
```

```python
# remove last column
df_drop_last_col = df_drop_last_row.drop([3], axis=1)
df_drop_last_col = df_drop_last_col.reset_index()
df_drop_last_col.head()
```

```python
# elevate first row to headers
new_header = df_drop_last_col.iloc[0]
new_df = df_drop_last_col[1:]
new_df.columns = new_header
new_df.head()
```

```python
# remove first column and reset index
new_df = new_df.drop([1],axis=1)
new_df.head()
```

```python
# reset table index
new_df = new_df.reset_index()
new_df = new_df[["Franchise","Latitude","Longitude"]]
new_df.head()
```

```python
# drop new york giants (duplicate) since the Jets already play in New York
# drop all the indoor stadiums
```

```python
locations = new_df.drop([new_df.index[3],new_df.index[9],new_df.index[20],new_df.index[19],new_df.index[17],new_df.index[11],new_df.index[12]])
locations
```

```python
# replace New York Jets with just New York
locations = locations.replace(to_replace="New York Jets", value="New York")
```

```python
# replace Saint Louis with just St. Louis
locations = locations.replace(to_replace="Saint Louis", value="St. Louis")
```

```python
# add 'Bay' to Tampa
locations = locations.replace(to_replace="Tampa", value="Tampa Bay")
```

```python
locations = locations.reset_index()
locations
```

```python
# remove the first column index as it is no longer of any value
locations = locations[["Franchise","Latitude","Longitude"]]
locations
```

```python
locations["Latitude"] = locations["Latitude"].str.rsplit(n=2, expand=True)
locations
```

```python
locations["Longitude"] = locations["Longitude"].str.rsplit(n=2, expand=True)
locations
```

```python
locations["Latitude"] = pd.to_numeric(locations["Latitude"], downcast="float")
locations["Longitude"] = pd.to_numeric(locations["Longitude"], downcast="float")
locations.dtypes
```

```python
locations["Longitude"] = locations["Longitude"].apply(lambda x: x*-1)
locations
```

```python
locations_dict = locations.to_dict(orient='series')
```

```python
print(locations_dict)
```

```python
# export table to csv
locations.to_csv("NFL_Locations.csv",encoding="utf-8",index=False)
```

```python
city= ['Chicago', 'Green Bay', 'Arizona', 'Baltimore', 'Buffalo', 
      'Carolina', 'Cincinnati','Cleveland', 'Denver', 
      'Indianapolis', 'Jacksonville', 'Kansas City',
      'Miami', 'New England', 'New York', 'Oakland', 'Philadelphia',
      'Pittsburgh', 'St. Louis', 'San Diego', 'San Francisco', 
      'Seattle', 'Tampa Bay', 'Tennessee', 'Washington']
```

```python
stadium_list = ['Soldier Field','Lambeau Field','State Farm Stadium','M&T Bank Stadium','New Era Field',
           'Bank of America Stadium','Paul Brown Stadium','FirstEnergy Stadium','Empower Field at Mile High',
           'Lucas Oil Stadium','TIAA Bank Field','Arrowhead Stadium',
          'Hard Rock Stadium','Gillette Stadium','MetLife Stadium','Oakland Coliseum(*)','Lincoln Financial Field',
          'Heinz Field','Edward Jones Dome','Qualcomm Stadium','Levis Stadium',
          'Lumen Field','Raymond James Stadium','Nissan Stadium','FedEx Field']
```

```python
stadium_dict = {'City':city,
                'Stadium':stadium_list
               }
stadium_dict
stadium_df = pd.DataFrame(stadium_dict)
```

```python
merged_stadium = pd.merge(left=locations, right=stadium_df, how='left', left_on="Franchise", right_on="City")
merged_stadium
```

```python
stadium_locations = merged_stadium[["City","Stadium","Latitude","Longitude"]]
```

```python
features = stadium_locations.apply(
    lambda row: Feature(geometry=Point((float(row['Latitude']), float(row['Longitude'])))),
    axis=1).tolist()

# all the other columns used as properties
properties = stadium_locations.drop(['Latitude', 'Longitude'], axis=1).to_dict('records')

# whole geojson object
feature_collection = FeatureCollection(features=features, properties=properties)
```

```python
with open('Resources/locations.geojson', 'w', encoding='utf-8') as f:
    json.dump(feature_collection, f, ensure_ascii=False)
```
