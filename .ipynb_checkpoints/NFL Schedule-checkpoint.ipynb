```python
# Import Libraries
import pandas as pd
from datetime import datetime
```

```python
# import and set up CSV
schedule_csv = "Resources/NFLSchedule.csv"

full_schedule = pd.read_csv(schedule_csv)
full_schedule.head()
```

```python
# remove columns
removed_col_df = full_schedule[['Year','Week','Date','Time','Winner/tie','HomeTeam']]
removed_col_df.head()
```

```python
full_date_df = removed_col_df
full_date_df["Full Date"] = full_schedule['Year'].map(str) + '-' + full_schedule["Date"].map(str)
full_date_df.head()
```

```python
full_date_df["Full Date"] = pd.to_datetime(full_date_df["Full Date"], format="%Y-%B %d")
```

```python
full_date_df.head()
```

```python
# filter for years  2008 - 2015 since API historical can only go back that far
year_filter_df = full_date_df[(full_date_df['Year']> 2007) & (full_date_df['Year'] <= 2015)]
year_filter_df.head()
```

```python
#year_filter_df["Time"] = year_filter_df["Time"].astype(str)
```

```python
city_split = year_filter_df
city_split[["City","Home Team"]] = year_filter_df["HomeTeam"].str.rsplit(n=1, expand=True)
city_split[["Winning City","Winner"]] = year_filter_df["Winner/tie"].str.rsplit(n=1, expand=True)
city_split.head()
```

```python
city_split["New Time"] = city_split["Time"].str.split(":", expand=True)
city_split.head()
```

```python
setup_df = city_split[["Full Date","New Time",'Week',"City","Home Team","Winner"]]
setup_df.head()
```

```python
setup_df.count()
```

```python
indoor_city = ["Dallas","New Orleans", "Atlanta","Minnesota","Detroit","Houston"]
```

```python
remove_indoor = setup_df[~setup_df['City'].isin(indoor_city)]
remove_indoor.head()
```

```python
remove_indoor.count()
```

```python
# ensure there are no indoor arenas by checking csv
#remove_indoor.to_csv("test.csv",encoding="utf-8",index=False)
```

```python
remove_indoor = remove_indoor.reset_index()
remove_indoor
```

```python
# fix time to display in hhmm (1am is 100, 3pm is 1500)
```

```python
final_table = remove_indoor[["Full Date","New Time",'Week',"City","Home Team","Winner"]]
final_table.head()
```

```python
final_table.to_csv("NFL_Schedule.csv",encoding="utf-8",index=False)
```
